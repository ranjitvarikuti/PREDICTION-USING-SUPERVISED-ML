Importing libraries and data

```python
import pandas as pnd
import numpy as nmp
import matplotlib.pyplot as pplt

#Importing necessary libraries to invoke functions defined under these.
```

```python
url = "http://bit.ly/w-data"  
r_data = pnd.read_csv(url)  #importing data into python script from remote link.
print("Data imported successfully")

r_data.head()  #head function shows the data for first 5 rows.
```

Study the data by plotting

```python
#Scatter Plot between number of hours studied and Scores scored by students
pplt.scatter(r_data['Hours'], r_data['Scores'])
pplt.title('Hours vs Percentages')
pplt.xlabel('Hours Studied')
pplt.ylabel('Percentage')
pplt.show()
```

```python
#To select specific rows and columns from dataset.
x = r_data.iloc[:, :-1].values
y = r_data.iloc[:, 1].values
print(x)
```

TRAINING OF DATASET

```python
#We will now divide the data set into train and test using built-in function train_test_split
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, 
                            test_size = 1/5, random_state = 0)
```

```python
# Now we need to train the model by 4/5 of data using built-in LinearRegression
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(x_train, y_train)
print('Done Training')
```

```python
# Regression line
line = regressor.coef_*x+regressor.intercept_

pplt.scatter(x, y)
pplt.plot(x, line, color = 'Orange');
pplt.show()
# We can see that we have positive correlation between hours and scores,
# That is with increase in number of study hours, there will be increase in scores.
```

PREDICTION

```python
#Prediction on 1/5 portion of dataset using predict function.
print(x_test)
y_pred = regressor.predict(x_test)
```

```python
# To compare the results of actual scores and predicted scores we form dataframe.
df = pnd.DataFrame({'Actual': y_test, 'Predicted': y_pred})
df
```

```python
# We can predict scores for our own new data.
hours=[9.25]
my_pred = regressor.predict([hours])
print("Number of Hours = {}".format(hours))
print("Predicted Score = {}".format(my_pred))
```

So, for a student who studies for about 9.25 hours can score 93.7% approximately.

ERROR ANALYSIS

```python
# Error occured in actual scores and predicted scores.
from sklearn import metrics
print('Mean Absolute Error:',
     metrics.mean_absolute_error(y_test, y_pred))
```
